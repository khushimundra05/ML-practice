```python
#using numpy 
import numpy as np
import matplotlib.pyplot as plt

x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]

plt.scatter(x,y)
plt.show() #see what the data points look like
```

```python
#clearly, non linear
#poly regression using numpy
import numpy
from sklearn.metrics import r2_score

mymodel=numpy.poly1d(numpy.polyfit(x,y,3))
myline=numpy.linspace(1,22,100) #generate an array of 100 evenly spaced values between 1 and 22

plt.scatter(x,y)
plt.plot(myline,mymodel(myline)) #values_to_be_used, line eqation
plt.show()

#evaluate using r2
print(r2_score(y,mymodel(x))) #compare model generated by mymodel vs actual y values
```
